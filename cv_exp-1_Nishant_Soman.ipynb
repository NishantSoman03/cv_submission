```python
import numpy as np

image = cv2.imread('logo.jpg')

height, width = image.shape[:2]

angle = 45  # Change this value to your desired rotation angle

rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)

rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
```

```python
plt.subplot(2,1,1)
plt.imshow(image)
plt.title('orignal')
plt.axis('off')  # Hide axes
plt.show()

plt.subplot(2,1,2)
plt.imshow(rotated_image)
plt.title('rotated')
plt.axis('off')  # Hide axes
plt.show()
```

```python
result1 = cv2.addWeighted(gray_img1, 0.5, gray_img2, 0.5, 0)

plt.imshow(result1, cmap='gray')
plt.title('addweighted')
plt.axis('off')  # Hide axes
plt.show()
```

```python
dx = 50  # shift along the x-axis
dy = 30  # shift along the y-axis
translation_matrix = np.float32([[1, 0, dx], [0, 1, dy]])

# Apply the translation to the image
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))

plt.subplot(2,1,1);
plt.imshow(translated_image)
plt.title('Og image')
plt.axis('off')  # Hide axes
plt.show()

plt.subplot(2,1,2)
plt.imshow(translated_image)
plt.title('translated')
plt.axis('off')  # Hide axes
plt.show()
```

```python
scaling_factor_x = 1.5  # Scaling factor along the x-axis
scaling_factor_y = 0.8  # Scaling factor along the y-axis

scaled_image = cv2.resize(image, None, fx=scaling_factor_x, fy=scaling_factor_y)

plt.subplot(2,1,1);
plt.imshow(image)
plt.title('Og image')
plt.axis('off')  # Hide axes
plt.show()

plt.subplot(2,1,2)
plt.imshow(scaled_image)
plt.title('scaled')
plt.axis('off')  # Hide axes
plt.show()
```
